```python
import matplotlib.pyplot as plt

# Top-level categories
main_categories = {
    "Cats": 12499,
    "Dogs": 12499,
    "Others": 12443
}

# Breakdown of "Others"
others_distribution = {
    "Flowers": 733,
    "Chair": 1500,
    "Fridge": 1500,
    "Table": 1931,
    "TV": 1500,
    "Human Faces": 2000,
    "Donkey": 499,
    "Horse": 480,
    "Zebra": 300,
    "Bird": 2000
}

# ---- Plot 1: Bar Chart for Top-level ----
plt.figure(figsize=(10,4))
plt.bar(main_categories.keys(), main_categories.values(), color=['orange', 'green', 'blue'])
plt.xlabel("Categories")
plt.ylabel("Number of Images")
plt.title("Image Dataset Distribution - Top Level")
plt.show()

# ---- Plot 2: Pie Chart for Top-level ----
plt.figure(figsize=(6,6))
plt.pie(main_categories.values(), labels=main_categories.keys(), autopct='%1.1f%%', startangle=90)
plt.title("Image Dataset Distribution - Top Level (Pie)")
plt.show()

# ---- Plot 3: Bar Chart for 'Others' breakdown ----
plt.figure(figsize=(12,5))
plt.bar(others_distribution.keys(), others_distribution.values(), color='skyblue')
plt.xlabel("Other Categories")
plt.ylabel("Number of Images")
plt.title("Breakdown of 'Others' Category")
plt.xticks(rotation=45)
plt.show()

# ---- Plot 4: Pie Chart for 'Others' breakdown ----
plt.figure(figsize=(7,7))
plt.pie(others_distribution.values(), labels=others_distribution.keys(), autopct='%1.1f%%', startangle=90)
plt.title("Breakdown of 'Others' Category (Pie)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ["Cats", "Dogs", "Others"]
train = [10624, 10624, 10576]
val = [1875, 1875, 1867]

x = np.arange(len(categories))  # x locations
width = 0.35  # width of the bars

# Plot
plt.figure(figsize=(8,5))
plt.bar(x - width/2, train, width, label="Train Set", color="skyblue")
plt.bar(x + width/2, val, width, label="Validation Set", color="orange")

# Labels
plt.xticks(x, categories)
plt.ylabel("Number of Images")
plt.title("Train vs Validation Set Distribution")
plt.legend()

plt.tight_layout()
plt.show()
```
